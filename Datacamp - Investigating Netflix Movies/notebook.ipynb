```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("netflix_data.csv")
df.head()
```

```python
#Subsetting the data frame to show only movies released between 1990 and 1999
Mv_90s = df[df["type"] == "Movie"]
subset = Mv_90s["release_year"].between(1990, 1999)
Mv_90s = Mv_90s[subset]
Mv_90s["release_year"].value_counts()
```

```python
# Creating a histogram to visualize the number of movie releases across the 90s decade
plt.hist(Mv_90s["release_year"])
plt.xlabel("Year")
plt.ylabel("Count of movies")
plt.title("Distribution of movies release in the 90s")
plt.show()
```

```python
#Checking the most common movie duration in the 1990s
duration = Mv_90s["duration"].mode().sum()
duration
```

Here, we can observe that the most frequent movie duration in the 1990s is 94 minutes. 

```python
#Confirming the data type of the variable "duration"
duration.dtype
```

```python
#Checking out the different genres of movies released in the 1990s
Mv_90s["genre"].value_counts()
```

```python
#Visualizing the genres in a bar plot
Mv_90s["genre"].value_counts().plot(kind= "bar", figsize= (16,8), rot= 0)
plt.xlabel("Genre")
plt.ylabel("Number of movies")
plt.title("Genres of movies released in the 1990s")
```

```python
#Subsetting the short action movies with duration less than 90 minutes
short_action_movies = (Mv_90s["genre"] == "Action") & (Mv_90s["duration"] < 90)
short_action_movies
```

```python
#Verifying the number of movies with short duration
short_movie_count = short_action_movies.sum()
short_movie_count
```

This result reveals that Netflix's catalogue contains only seven short action movies released in the 1990s.
